# Tool Creation

```python
from langchain_core.tools import tool
from langchain_ollama import ChatOllama
from langchain_core.messages import HumanMessage

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two number"""
    return a * b

print(multiply.invoke({"a": 3, "b": 5}))
print(multiply.description)
print(multiply.name)
print(multiply.args)
print(multiply.args_schema)
```

# Tool binding

```python
llm = ChatOllama(model="qwen2.5")

model = llm.bind_tools([multiply])

print(model)
```

```python

response = model.invoke("Hey how are you ")

print(response)
print(response.tool_calls)
```

# Tool calling

```python
prompt = HumanMessage("Can you multiply 3 with 10 ?")
messages = [prompt]
messages
```

```python

response = model.invoke([prompt])
print(response)
print(response.tool_calls)
messages.append(response)
messages
```

# Tool Execution

```python

result = multiply.invoke(response.tool_calls[0]['args'])
result
```

```python
result = multiply.invoke(response.tool_calls[0])
print(result)
messages.append(result)
```
